```python
ts_length = 100
ϵ_values = []

for i in range(ts_length):
    e = np.random.randn()
    ϵ_values.append(e)

plt.plot(ϵ_values)
plt.show()
```

```python
ts_length = 100
ϵ_values = []
i = 0

while i < ts_length:
    e = np.random.randn()
    ϵ_values.append(e)
    i = i + 1

plt.plot(ϵ_values)
plt.show()
```

```python
r = 0.025
T = 50
b = np.empty(T+1)
b[0] = 10

for t in range(T):
    b[t+1] = (1 + r) * b[t]

plt.plot(b, label='bank balance')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
alpha = 0.9
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    x[t+1] = (alpha * x[t]) + np.random.randn()

plt.plot(x)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
alphas = [0, 0.8, 0.98]
x = np.empty(T+1)
x[0] = 0

for alpha in alphas:
    for t in range(T):
        x[t+1] = (alpha * x[t]) + np.random.randn()
    plt.plot(x, label=f'alpha = {alpha}')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
alpha = 0.9
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    x[t+1] = alpha * np.abs(x[t]) + np.random.randn()

plt.plot(x)
plt.show()
```

```python
import numpy as np

n = 1000000
count = 0

for i in range(n):
    u, v = np.random.uniform(), np.random.uniform()

    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)

    if d < 0.5:
        count += 1

area_estimate = count / n
print(area_estimate * 4)
```
